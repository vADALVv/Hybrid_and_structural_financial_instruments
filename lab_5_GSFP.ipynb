## Аппроксимация геометрическим средним (Gentel)

```python
# аппроксимация корзины опционов геометрическим средним
def geometric_average_approximation_basket_call(S, K, T, w, r, rho, sigma):
  F = S * np.exp(r * T)
  a = w * F / np.sum(w * F)

  m_tilde = np.log(np.sum(w * F)) - 0.5 * np.sum(a * sigma ** 2 * T)
  v_tilde = np.sqrt((a * sigma).T @ rho @ (a * sigma) * T)

  K_star =  K - np.sum(w * F) + np.exp(m_tilde + 0.5 * v_tilde ** 2)

  d1 = (m_tilde - np.log(K_star) + v_tilde ** 2) / v_tilde
  d2 = d1 - v_tilde

  basket_call = np.exp(-r * T) * (np.exp(m_tilde + 0.5 * v_tilde ** 2) * norm.cdf(d1) - K_star * norm.cdf(d2))

  return basket_call
```

## Метод условного математического ожидания (Beisser)

```python
def get_z_star(x, S, K, r, q, sigma, T, w, rho):
    res = 0
    a = S * w * sigma
    sqrt_T = np.sqrt(T)
    sigmaz = np.sqrt(a.T @ rho @ a)
    for i in range(0, len(w)):
        m = np.sum(a * rho[i]) / sigmaz**2 * sqrt_T
        exp_S_Z = S[i] * np.exp(sigma[i] * m * x - 1/2 * sigma[i]**2 * m**2 * sigmaz**2 + (r - q[i]) * T)
        res += w[i] * exp_S_Z
    return (res - K)**2

def Beisser_approximation_basket_call(S, K, T, w, r, q, sigma):
  option = 0
  a = S * w * sigma
  sigmaz = np.sqrt(a.T @ rho @ a)
  sqrt_T = np.sqrt(T)

  z_star = minimize(get_z_star, 0, args=(S, K, r, q, sigma, T, w, rho)).x[0]
  for i in range(0, len(w)):
    m = np.sum(a * rho[i]) / sigmaz**2 * sqrt_T
    sigma_star = sigma[i] * m * sigmaz / np.sqrt(T)
    K_star = S[i] * np.exp(sigma[i] * m * z_star - 1/2 * sigma[i]**2 * m**2 * sigmaz**2 + (r - q[i]) * T)
    S_star = S[i] * np.exp((r - q[i] - 1/2 * sigma_star**2) * T + sigma_star * np.sqrt(T) / sigmaz * z_star)
    d1 = (np.log(S[i] / K_star) + (r - q[i] + 1/2 * sigma_star**2) * T) / (sigma_star * np.sqrt(T))
    d2 = d1 - sigma_star * np.sqrt(T)
    option += w[i] * (np.exp(-q[i] * T) * S[i] * norm.cdf(d1) - np.exp(-r * T) * K_star * norm.cdf(d2))
    #print(f"d1 = {d1}, d2 = {d2}, z_star = {z_star}, K_star = {K_star}, S_star = {S_star}, sigma_star = {sigma_star}")
  return option
```

```python
Beisser_approximation_basket_call(S, K, T, w, r, q, sigma)
```

### Метод Джу через разложения в ряд Тейлора (Ju)

Статья [(Ju, 2002)](https://drive.google.com/file/d/1JJuZhRiCUGf6euyfju1Dn5aRP46-3iry/view?usp=drive_link)

```python
def ju_taylor_approximation_call(S, K, T, sigma, w, rho):
    # корреляционная матрица
    rho_matrix = np.full((len(S), len(S)), rho) + np.eye(len(S)) * (1 - rho)

    # расчет взвешенной волатильности корзины
    sigma_squared = sigma ** 2
    basket_variance = np.dot(w, np.dot(rho_matrix * np.outer(sigma, sigma), w))
    basket_sigma = np.sqrt(basket_variance)  # волатильность корзины

    # расчет текущей цены корзины активов
    basket_price = np.sum(w * S)

    d1 = (np.log(basket_price / K) + (0.5 * basket_sigma**2) * T) / (basket_sigma * np.sqrt(T))
    d2 = d1 - basket_sigma * np.sqrt(T)

    call_price = basket_price * norm.cdf(d1) - K * norm.cdf(d2)

    return call_price

```

## Метод моментов (Log-normal moment matching, Levy’s matching)

```python
def Levy_approximation_basket_call(S, K, T, w, r, rho, sigma):
  F = S * np.exp(r * T)
  M = np.dot(w, F)
  V2 = 0
  for i in range(0, len(w)):
    for j in range(0, len(w)):
      V2 += w[i] * w[j] * F[i] * F[j] * np.exp(sigma[i] * sigma[j] * rho[i][j] * T)
  m = 2 * np.log(M) - 1/2 * np.log(V2)
  v = np.sqrt(np.log(V2) - 2 * np.log(M))
  d1 = (m - np.log(K) + v**2) / v
  d2 = d1 - v
  basket_call = np.exp(-r * T) * (M * norm.cdf(d1) - K * norm.cdf(d2))
  return basket_call
```

```python
Levy_approximation_basket_call(S, K, T, w, r, rho, sigma)
```

## Аппроксимация через обратное геометрическое распределение (MP-RG)

```python
def Milevsky_Posner_RG_approximation_basket_call(S, K, T, w, r, rho, sigma):
  F = S * np.exp(r * T)
  M = np.dot(w, F)
  V2 = 0
  for i in range(0, len(w)):
    for j in range(0, len(w)):
      V2 += w[i] * w[j] * F[i] * F[j] * np.exp(sigma[i] * sigma[j] * rho[i][j] * T)
  alpha = (2 * V2  - M**2) / (V2 - M**2)
  beta = (V2 - M**2) / (V2 * M)
  g1 = gamma.cdf(1/K, a = alpha - 1, scale = beta)
  g2 = gamma.cdf(1/K, a = alpha, scale = beta)
  basket_call = np.exp(-r * T) * (M * g1 - K * g2)
  return basket_call
```

```python
import numpy as np
from scipy.stats import norm, skew, kurtosis

def calculate_skewness_kurtosis(prices): # prices: матрица исторических цен (строки — даты, столбцы — активы)

    returns = np.diff(prices, axis=0) / prices[:-1] 
    skewness = skew(returns, axis=0) 
    kurtosis_vals = kurtosis(returns, axis=0, fisher=False)  
    return skewness, kurtosis_vals


def Milevsky_Posner_4M_approximation_basket_call(S, K, T, w, r, rho, sigma, prices):

    n = len(S) 

    # асимметрия и эксцесс
    skewness, kurtosis_vals = calculate_skewness_kurtosis(prices)
    avg_skewness = np.mean(skewness) 
    avg_kurtosis = np.mean(kurtosis_vals)  

    # мат ожидание 
    mu = np.dot(w, S) * np.exp(r * T)

    # дисперсия 
    sigma_matrix = np.outer(sigma, sigma) * (rho * (np.ones((n, n)) - np.eye(n)) + np.eye(n))
    sigma_squared = np.dot(w, np.dot(sigma_matrix, w))
    sigma_basket = np.sqrt(sigma_squared)

    # параметры распределения Джонсона для SPD
    if avg_kurtosis > (3 + avg_skewness**2):  # ПType I
        # Type I (Shifted Lognormal)
        w_param = ((8 + 4 * avg_skewness**2 + np.sqrt(16 * avg_skewness**2 + 64)) / 4) ** (1 / 3)
        b = 1 / np.sqrt(np.log(w_param))
        a = 0.5 * b * np.log((w_param - 1) / avg_skewness**2)
        d = sigma_basket
        c = mu - d * np.exp((0.5 / b) - a / b)
    else:
        # Type II (Sinh-inverse hyperbolic)
        w_param = np.sqrt(2 * avg_kurtosis - 2)
        b = 1 / np.sqrt(np.log(w_param))
        a = 0
        d = sigma_basket
        c = mu - d * np.sinh(a / b)

    d1 = (np.log(mu / K) + 0.5 * sigma_basket**2 * T) / (sigma_basket * np.sqrt(T))
    d2 = d1 - sigma_basket * np.sqrt(T)

    basket_call = mu * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)
    return basket_call

```

## Метод Монте-Карло

```python
def MC_basket_call(S, K, T, w, r, q, rho, sigma, N_mc):
  n = len(w)
  COV = sigma.T * rho * sigma
  W = np.random.multivariate_normal(np.zeros(n), COV, size=N_mc) * np.sqrt(T)
  B_T = np.zeros(N_mc)
  S0 = 100
  for k in range(0, n):
      B_T += S[k] * np.exp((r - q[k] - sigma[k]**2 / 2) * T + W[:,k]) * w[k]
  basket_call = np.mean(np.maximum(B_T - K, 0)) * np.exp(-r * T)
  return basket_call
```

```python
# МС оценка для кол и пут опционов
n_steps = 5000
n_sims = 10000
dt = T / (n_steps - 1)
COV = sigma.T * rho * sigma

S = np.zeros((n_sims, 4))
S[...] = F

for i in tqdm(range(n_steps - 1)):
    dW = np.random.multivariate_normal(np.zeros(4), COV, size=n_sims) * np.sqrt(dt)
    dS = r * S * dt + S * dW
    S += dS

portfolio_values = S @ w

call_price_mc = np.maximum(portfolio_values - K, 0).mean()
put_price_mc = np.maximum(K - portfolio_values, 0).mean()

print('Call price by Monte-Carlo:', call_price_mc)
print('Put price by Monte-Carlo:', put_price_mc)
```

```python
def MC_basket_call(S, K, T, w, r, q, rho, sigma, N_mc):
    n = len(w)
    COV = np.outer(sigma, sigma) * rho
    W = np.random.multivariate_normal(np.zeros(n), COV, size=N_mc) * np.sqrt(T)
    B_T = np.zeros(N_mc)
    for k in range(n):
        B_T += S[k] * np.exp((r - q[k] - sigma[k]**2 / 2) * T + W[:, k]) * w[k]
    basket_call = np.mean(np.maximum(B_T - K, 0)) * np.exp(-r * T)
    return basket_call

def MC_basket_put(S, K, T, w, r, q, rho, sigma, N_mc):
    n = len(w)
    COV = np.outer(sigma, sigma) * rho
    W = np.random.multivariate_normal(np.zeros(n), COV, size=N_mc) * np.sqrt(T)
    B_T = np.zeros(N_mc)
    for k in range(n):
        B_T += S[k] * np.exp((r - q[k] - sigma[k]**2 / 2) * T + W[:, k]) * w[k]
    basket_put = np.mean(np.maximum(K - B_T, 0)) * np.exp(-r * T)
    return basket_put

def MC_basket_call_with_std(S, K, T, w, r, q, rho, sigma, N_mc):
    n = len(w)
    COV = np.outer(sigma, sigma) * rho
    W = np.random.multivariate_normal(np.zeros(n), COV, size=N_mc) * np.sqrt(T)
    B_T = np.zeros(N_mc)
    for k in range(n):
        B_T += S[k] * np.exp((r - q[k] - sigma[k]**2 / 2) * T + W[:, k]) * w[k]
    payoff = np.maximum(B_T - K, 0)
    mean = np.mean(payoff) * np.exp(-r * T)
    std = np.std(payoff) * np.exp(-r * T)
    return mean, std

def MC_basket_put_with_std(S, K, T, w, r, q, rho, sigma, N_mc):
    n = len(w)
    COV = np.outer(sigma, sigma) * rho
    W = np.random.multivariate_normal(np.zeros(n), COV, size=N_mc) * np.sqrt(T)
    B_T = np.zeros(N_mc)
    for k in range(n):
        B_T += S[k] * np.exp((r - q[k] - sigma[k]**2 / 2) * T + W[:, k]) * w[k]
    payoff = np.maximum(K - B_T, 0)
    mean = np.mean(payoff) * np.exp(-r * T)
    std = np.std(payoff) * np.exp(-r * T)
    return mean, std

def CI_MC(price, price_std, N_mc):
    lo = price - norm.ppf(0.975) * price_std / np.sqrt(N_mc)  # lower bound of confidence interval for the price
    hi = price + norm.ppf(0.975) * price_std / np.sqrt(N_mc)  # upper bound of confidence interval for the price
    print(f"mean basket price = {price}, ci: {lo} < price < {hi} with 95% prob")
```
